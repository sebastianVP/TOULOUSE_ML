### Importamos librerias -Aplicación practica de FAMD con Python


Primero, necesitamos instalar la librería prince que nos proporciona la implementación de FAMD en Python. Puedes hacerlo mediante la siguiente línea de código en tu terminal:



```python
pip install prince
```

### Cargamos datos de entrada -DATASET IRIS

```python
import pandas as pd
from prince import FAMD

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df.head()

df.shape
```

### Ahora, podemos aplicar el FAMD al conjunto de datos Iris:

##### EIGENVALUES SUMMARY

```python
famd = FAMD(
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    random_state=42,
    engine="sklearn",
    handle_unknown="error"  # same parameter as sklearn.preprocessing.OneHotEncoder
)
famd = famd.fit(df)
```

En este caso, estamos reduciendo las cuatro dimensiones originales del conjunto de datos Iris a dos nuevas dimensiones. Y a continuación podemos apreciar el porcentaje de variación atrapado por cada eigenvalue.

```python
famd.eigenvalues_summary
```

Coordenadas de cada fila y columna.

```python
famd.row_coordinates(df).shape
```

```python
famd.row_coordinates(df).head()
```

```python
famd.column_coordinates_,
```

```python
famd.plot(
    df,
    x_component=0,
    y_component=1
)
```

Contribución de cada fila y columna a cada componente.



```python
famd.row_contributions_.sort_values(0, ascending=False).head(5).style.format('{:.3%}')
```

```python
famd.column_contributions_.style.format('{:.0%}')
```
